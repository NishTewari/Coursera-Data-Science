### The following code below demonstrates a solution to the Fizzbuzz problem - Numbers divisble by 3 are "Fizz" & Numbers divisble by 5 are "buzz" & numbers divisble by both 3 and 5 are "Fizzbuzz"

```python
for num in range(1,50):
    string = ""
    if num % 5 == 0 and num % 3 == 0:
        string = string + "FizzBuzz"
    elif num % 3 == 0:
        string = string + "Fizz"
    elif num % 5 ==0:
        string = string + "Buzz"
    else:
        string = string + str(num)
    print(string, end=' ')
```
